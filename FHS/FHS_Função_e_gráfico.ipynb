Este notebook tem como objetivo gerar um gráfico para uma função qualquer.

```python
# Carregando bibliotecas necessárias
import numpy as np                # para trabalhar com vetores e matrizes
import matplotlib.pyplot as plt   # para gerar o gráfico
```

```python
# Definindo a função de deve dar origem ao gráfico. Exemplo: y = x^2 + 4x - 5
def funcao(x):
    return x**2 + 4*x - 5
```

```python
# Recebe a faixa de valores de x. Ex: para gerar o gráfico de -5 até +5, para o valor de x variando de 2 em 2, os valores informados devem ser: -5, 5 e 2. Neste caso, os valores de x usados serão -5, -3, -1, 1, 3 e 5.
x_min = float(input("Digite o valor mínimo de x: "))
x_max = float(input("Digite o valor máximo de x: "))
x_intervalo = float(input("Digite o intervalo de variação de x: "))

# Gera uma lista de valores de x dentro da faixa especificada
x = np.arange(x_min, x_max, x_intervalo)

# Calcula os valores de y para cada valor de x
y = funcao(x)

# Gera o gráfico
plt.plot(x, y, label='y = x^2 + 4x - 5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de y em função de x')
plt.legend()
plt.grid(True)
plt.show()
```
